```python
# プログラムの先頭において，そのプログラムで使うライブラリをimport文を用いて呼び出す
import numpy as np                # numpy: 数値計算のライブラリ
import matplotlib.pyplot as plt   # matplotlib: グラフ描画，画像表示のためのライブラリ
import matplotlib.animation as animation
from IPython import display
```

```python
um_pix = 0.1036

fig, ax = plt.subplots(figsize=(10,10))

image0 = plt.imread('original/image0000.tif')
mappable = ax.imshow(image0, cmap='gray', extent=[0, 2048*um_pix, 0, 2048*um_pix]) # グラフを指示する関数から返り値を受け取り
cbar = fig.colorbar(mappable) # カラーバーへ入力する
cbar.set_label('Intensity') # カラーバーのラベル

def update(idImage):
    img = plt.imread(f'original/image{idImage:04}.tif')
    ax.imshow(img, cmap='gray', extent=[0, 2048*um_pix, 0, 2048*um_pix])

ani = animation.FuncAnimation(fig, update, frames=10, interval=200)

html = display.HTML(ani.to_jshtml())
display.display(html)

plt.close()
```

```python
ani.save('test.mp4')
```

```python
from matplotlib_scalebar.scalebar import ScaleBar

fig, ax = plt.subplots(figsize=(10,10))

image0 = plt.imread('original/image0000.tif')
mappable = ax.imshow(image0, cmap='gray', extent=[0, 2048*um_pix, 0, 2048*um_pix]) # グラフを指示する関数から返り値を受け取り
cbar = fig.colorbar(mappable) # カラーバーへ入力する
cbar.set_label('Intensity') # カラーバーのラベル

scalebar = ScaleBar(1, 'um', fixed_value=60, location='lower right')
ax.add_artist(scalebar)

plt.show()
```
